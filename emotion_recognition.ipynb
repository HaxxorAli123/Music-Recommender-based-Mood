```python
from ultralytics import YOLO
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os
import zipfile
```

```python
print(os.getcwd())
```

```python
zip_file_path = "mood-detection.v4i.yolov11.zip"
extract_to = 'datasets'
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_to)
```

```python
PRETRAINED_MODEL_PATH = 'yolo11n.pt'
DATASET_YAML_PATH = r"C:\Users\User\OneDrive\Desktop\Final_Caps\syed_capstone_vision_capstone_5\datasets\data.yaml"
TRAINED_MODEL_PATH = r"C:\Users\User\OneDrive\Desktop\Final_Caps\syed_computer_vision_capstone_5\runs\detect\train\weights\epoch20.pt"
```

```python
emotion_detection_model = YOLO(PRETRAINED_MODEL_PATH)
```

```python
emotion_detection_model.train(
    data=DATASET_YAML_PATH,  # Path to dataset YAML
    epochs=25,               # Training for 20 epochs
    batch=8,                 # Batch size
    imgsz=640,               # Image size
    workers=2,               # Number of DataLoader workers
    optimizer="SGD",         # Optimizer for training
    patience=10,             # Early stopping patience
    save_period=5,           # Save weights every 5 epochs
    val=True,                # Evaluate on validation set
    verbose=True             # Enable detailed logs
)
```

```python
best_emotion_detection_model = YOLO(TRAINED_MODEL_PATH)
```

```python
# Test the model on a sample image
sample_image_path = 'sample.jpg'
sample_result = best_emotion_detection_model.predict(sample_image_path)

# Check if predictions exist and display the results
if sample_result and len(sample_result) > 0 and sample_result[0] is not None:
    img = sample_result[0].plot()
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert color for display
    plt.imshow(img)
    plt.xticks([])
    plt.yticks([])
    plt.show()
else:
    print("No predictions were made for the sample image.")
```

```python
fined_results = best_emotion_detection_model.predict(source=0, show=True, stream=True)
```

```python
for result in fined_results:
    print(result.boxes)  # Print bounding boxes
    print(result.names)  # Print detected class names
    if cv2.waitKey(1) == ord('q'):
        break

cv2.destroyAllWindows()
```
